```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD
from torch.utils.data import TensorDataset, DataLoader
import torchvision.transforms.functional as TF
import torchvision
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
import numpy as np
```

### Importing the training data

```python
label_translations = np.load("class_names.npy", allow_pickle=True).item()
```

```python
train_paths = pd.read_csv('train_images.csv')
images = []
labels = []
print('Importing and resizing images...')
newsize = (256, 256)
for path, label, i in zip(train_paths['image_path'], train_paths['label'], range(len(train_paths))):
    if round((i / len(train_paths)) * 100) % 1 == 0:
        print(f'{round((i / len(train_paths)) * 100)}% complete', end="\r")
    path = 'train_images' + path
    img = Image.open(path).convert('RGB').resize(newsize)
    images.append(img)
    labels.append(label)
```

```python
X = np.array(images)
# channel first format
#X = np.transpose(X, (0, 3, 1, 2))
Y = np.array(labels)
```

### Some visualisations

```python
# plotting example images
np.random.seed(22)
fig = plt.figure(figsize=(12, 6))
rows, cols = 2, 4
for i in range(1, rows * cols + 1):
    random_index = np.random.randint(0, len(X), size=[1]).item()
    img, label = X[random_index], Y[random_index]
    fig.add_subplot(rows, cols, i)
    plt.imshow(img.squeeze(), cmap="gray")
    plt.title(list(label_translations.keys())[list(label_translations.values()).index(label)][:])
    plt.axis(False)
```
